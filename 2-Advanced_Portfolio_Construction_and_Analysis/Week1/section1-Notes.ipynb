## Changing Asset Management Paradigm

Investing in fixed income markets has undergone a big transformation in recent years. For starters, factor investing—originally popularized by index providers like MSCI and commonly found in equity portfolios—is now spreading into fixed income. Sometimes called "smart beta" or "systematic investing," these quantitative ("quant") investment strategies seek to generate active returns by using rules-based factors (like value, momentum and quality) that identify persistent quantifiable drivers of excess returns. These quant-oriented approaches belong to a new investment category (shown in Exhibit 1) that theoretically sits in between passive index strategies offering market beta and active managers who deliver "alpha" (i.e., excess returns not explained by the overall market’s rate of return, or "beta"), the usual old school asset management theory.

## Nutrient Analogy

Factor investing:  

* A factor is a variable which influences the **returns of assets**
* It represents commonality in the returns. Somehting *outside the individual asset*
* Exposure to factor risk over the long run yields a reward. **The risk Premium*

Types of factors:

* _Macro factors_: Growth, industry growth, inflation, ...
* _Statistical factors_: something extracted from the data that may or may not be identifiable
* _Intrinsic factors or style factors_: value-growth, momentum, low volatility, ...



## Factor Models and  the CAPM

* A factor model is nothing more than a decomposition of the returns $R$ of an asset into a set of factor premia:
$$R_i = \beta_1 f_1 + \beta_2 f_2 + \dots + \beta_i f_i + \alpha + \epsilon$$
* Factor premia are returns that you  get in excange for exposing yourself to the factor
* Epsilon refers to the error term, and is a measure of what might be in the data that is not captured by the factor model 
* Meaning: If you can decompose the observed returns o an asset into the returns of the drivers of these factors or the factor premia, then you have **a factor model**

<div class="alert alert-block alert-warning">
So whenever you hear the word factor model, just in your head, think of it as a decomposition of the returns into a set of factor premium.
</div>

## CAPM

* The capital asset pricin model (CAPM) is the simplest and probably best known factor model. 
* It includes  more than a factor model. 
* The CAPM predicts that the alpha term is zero


![CAPM](../img/capm.png)

<div class="alert alert-block alert-info">
The excess return of an asset over the risk-free rate is nothing more than the excess return on the market over the risk-free rate, times the Beta of the stock.</div>

* So what you've done is you've decomposed, in a sense, the returns of an asset into one factor return, and that one factor is the return on the market

* The excess return of an asset is nothing more than some measure of the market's return multiplied by the exposure of that stock to the market's return which is what we call the Beta.  

* The factor in this case is the market. So if you have a high Beta stock, what that basically means is if the market's up, your stock's going up; and so on.
* Now, if you have a low Beta stock, let's imagine that you have a stock with a Beta of zero. It's going to be completely decorrelated from the market. It doesn't matter what the excess return of the market is going to be because the expected return from the stock is just zero times the excess return on the market. That means everything is going to be in that error term

<div class="alert alert-block alert-warning">
The CAPM's factor model  tells us that the returns of assets are just explained by one factor: the return on the market and the Beta.</div>

*Now, in reality, we know that doesn't really happen. We know that when we actually look at the numbers, we find that certain types of stocks outperform other types of stocks, the best-known probably and it's been known for a long time is value stocks. Value stocks outperform growth stocks. There have been other more weird anomalies as these have been called, for example, low-vol stocks, as a class, tend to outperform at least on a risk adjusted basis high-vol stocks. The exact opposite of what CAPM expects. This is what we call **anomalies**.*

<div class="alert alert-block alert-warning">

So, if you find some effect like this that is not explained by beta, then clearly it is an anomaly from the perspective of the CAPM, and isn't really well-explained by the CAPM.</div>

### Fama and French (1993)

**<span style="color:red">The Fama French model extends the CAPM by adding Value and Size as Factors.</span>**

<div class="alert alert-block alert-warning">

So they introduced these new factors in addition to the market factor. And they establish the fact that you get a better explanation of the returns of assets when you add these additional factors beyond just simply the market factor
